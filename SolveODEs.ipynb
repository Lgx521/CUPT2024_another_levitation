```python
from kit.functions import *
from kit.Constants import *
import matplotlib.pyplot as plt
import numpy as np
import scipy.integrate as s

path = '/Users/gansz/Downloads/CUPT2024/PythonAnalysis/plots/ODE_solver_plot.png'

def func(t, x):
    dx1 = x[1]
    dx2 = F(x[0]) * x[1] + F_m(x[0]) - g
    return [dx1, dx2]

x_f = [0.04539]*12000

t = np.linspace(0, 2, 2000)
t2 = np.linspace(0, 12, 12000)

# Initial conditions
x0 = [0.1, 0]

# ODE solver
x = s.odeint(func, x0, t, tfirst=True)
x2 = s.odeint(func, x0, t2, tfirst=True)

# Figure plot
plt.figure(figsize=(12,10))


plt.subplot(2,1,1)
plt.plot(t, x[:,0],lw='1.5',label='position')
plt.plot(t,x_f[0:2000:],lw='1',label='Steady state', color='darkorange')
plt.grid()
plt.legend()
plt.title('The initial distance between the two magnet is 0.1 m\n\nThe position-time graph of the falling magnet in 2 sec')
plt.ylabel('z / meter')

plt.subplot(2,1,2)
plt.plot(t2, x2[:,0],lw='0.5',label='position',color='g')
plt.plot(t2,x_f,lw='1',label='Steady state', color='darkorange')
plt.grid()
plt.legend()
plt.xlabel('time / second')
plt.ylabel('z / meter')
plt.title('The position-time graph of the falling magnet in 12 sec')

plt.savefig(path, dpi=500)
 
plt.show()
    
```
